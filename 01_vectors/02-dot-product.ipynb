:::{note} Definition: Dot Product
Note that you can add definition boxes (and choose how they're formatted) like this.
The dot product of $\vec u, \vec v \in \mathbb{R}^n$ is

$$\vec u \cdot \vec v = \sum_{i=1}^n u_i v_i$$
:::

### Geometric Interpretation

It connects algebra and geometry via

$$\vec u \cdot \vec v = \|\vec u\| \|\vec v\| \cos \theta$$

If the dot product is zero, the vectors are orthogonal.
